```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
```

```python
x = np.array(range(1,50))
x1 = 2
y1 = (0 * x) + 1
y2 = np.log(x)
y3 = x
y4 = x * np.log(x)
y5 = x**2
y6 = 2 ** x
y7 = x * scipy.special.factorial(x)
y8 = x ** x

plt.plot(x, y1, c="r", label="O(1)")
plt.plot(x, y2, c="b", label= "O(log n)")
plt.plot(x, y3, c="g", label="O(n)")
plt.plot(x, y4, c="y", label="O(n log n)")
plt.plot(x, y5, c="indigo", label="$O(n^2)$")
plt.plot(x, y6, c="pink", label="$O(2^n)$")
plt.plot(x, y7, c="orange", label="O(n!)")
plt.plot(x, y8, c="black", label="$O(n^n)$")
plt.title("Big O Notation")
plt.xlabel("Elements")
plt.ylabel("Operations")
plt.legend(loc="upper left")
plt.show()
```
