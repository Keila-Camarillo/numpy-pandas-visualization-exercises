```python
import pandas as pd
from pydataset import data
import numpy as np

np.random.seed(123)

students = ['Sally', 'Jane', 'Suzie', 'Billy', 'Ada', 'John', 'Thomas',
            'Marie', 'Albert', 'Richard', 'Isaac', 'Alan']

# randomly generate scores for each student for each subject
# note that all the values need to have the same length here
math_grades = np.random.randint(low=60, high=100, size=len(students))
english_grades = np.random.randint(low=60, high=100, size=len(students))
reading_grades = np.random.randint(low=60, high=100, size=len(students))

df = pd.DataFrame({'name': students,
                   'math': math_grades,
                   'english': english_grades,
                   'reading': reading_grades})

type(df)

```

```python
# 1. Copy the code from the lesson to create a dataframe full of student grades.
df
```

```python
# a. Create a column named passing_english that indicates whether each student has a passing grade in english.
df["passing_english"] = df.english >= 70
```

```python
df
```

```python
# b. Sort the english grades by the passing_english column. How are duplicates handled? indexes

df.sort_values(by=["passing_english"])
```

```python
# c. Sort the english grades first by passing_english and then by student name. 
# All the students that are failing english should be first, and within the students 
# that are failing english they should be ordered alphabetically. 
# The same should be true for the students passing english. (Hint: you can pass a list to the .sort_values method)
df.sort_values(by=["passing_english", "name"])
```

```python
########

# d. Sort the english grades first by passing_english, and then by the actual english grade, 
# similar to how we did in the last step.
df.sort_values(by=["name"]).sort_values(by=["passing_english", "english"], ascending=[True, True])
```

```python
# e. Calculate each students overall grade and add it as a column on the dataframe. 
# The overall grade is the average of the math, english, and reading grades.
df['overall'] = round(df[["math", "english", "reading"]].mean(axis=1), 2)
df
```

```python
# 2. Load the mpg dataset. Read the documentation for the dataset and use it for the following questions:
data('mpg', show_doc=True)

mpg = data('mpg') # load the dataset and store it in a variable
```

```python
# How many rows and columns are there?
mpg.shape
```

```python
# What are the data types of each column?
mpg.dtypes
```

```python
# Summarize the dataframe with .info and .describe
print(mpg.info)
print(mpg.describe())
```

```python
# Rename the cty column to city.
mpg = mpg.rename(columns={"cty": "city"})
```

```python
# Rename the hwy column to highway.
mpg = mpg.rename(columns={"hwy": "highway"})
```

```python
mpg
```

```python
# Do any cars have better city mileage than highway mileage? No
print(mpg.city > mpg.highway)
```

```python
# Create a column named mileage_difference this column should contain the difference between highway 
# and city mileage for each car.
mpg["mileage_difference"] = mpg.highway - mpg.city
```

```python
# Which car (or cars) has the highest mileage difference?
mpg[mpg.mileage_difference == mpg["mileage_difference"].max()]
```

```python
# Which compact class car has the lowest highway mileage? The best?
mpg[mpg['class'] == 'compact'].sort_values('highway').head(1) # worst
mpg[mpg['class'] == 'compact'].sort_values('highway', ascending=False).head(1) # best
```

```python
# Create a column named average_mileage that is the mean of the city and highway mileage.
mpg["average_mileage"] = mpg[["city", "highway"]].mean(axis=1)
```

```python
mpg
```

```python
# Which dodge car has the best average mileage? The worst?
mpg[mpg["manufacturer"] == 'dodge'].sort_values('average_mileage', ascending=False).head(1) # best mileage
mpg[mpg["manufacturer"] == 'dodge'].sort_values('average_mileage').head(3) # worst mileage
```

```python
# How many animals are hoppers that are above the median speed? What percentage is this?
mammals.speed.median()
mammals.loc[(mammals.hoppers == True) & (mammals.speed > mammals.speed.median())].shape[0] # number of animals
(mammals.loc[(mammals.hoppers == True) & (mammals.speed > mammals.speed.median())].shape[0]/mammals.shape[0]) * 100 # percentage
```
